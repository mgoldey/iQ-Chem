```python
print "This is code for a simple 1D scan which can be generalized for other paramter changes in Q-Chem"
def qrun(name,r,dist="foo"):
    nm=name+"_"+str(r)+".in"
    nmin=nm
    nmout=nm.replace(".in",".out")
    nmdir=nm.replace(".in",".dir")
    if os.path.isfile(nmout):
        op=file(nmout,'r').read()
        try:
            c=op.index("Total energy")
            return float(op[c:].splitlines()[0].split()[-1])
        except:
            print "Output not right"
            return
    xyz=file(name+".xyz","r").read().split('\n\n')[1]
    a=file(nm,'w')
    a.write("$molecule\n0 1\n")
    a.write(xyz.replace(dist,str(r)))
    a.write("\n$end\n")
    a.write(rem)
    a.close()
    os.popen("qchem -nt 12 "+nmin+" "+nmout+" "+nmdir).read()
    op=file(nmout,'r').read()
    try:
        c=op.index("Total energy")
        return float(op[c:].splitlines()[0].split()[-1])
    except:
        print "Something went wrong for ",nm
        return
def q_ser_run(name,r,dist="foo"):
    nm=name+"_"+str(r)+".in"
    nmin=nm
    nmout=nm.replace(".in",".out")
    nmdir=nm.replace(".in",".dir")
    if os.path.isfile(nmout):
        op=file(nmout,'r').read()
        try:
            c=op.index("Total energy")
            return float(op[c:].splitlines()[0].split()[-1])
        except:
            print "Output not right"
            return
    xyz=file(name+".xyz","r").read().split('\n\n')[1]
    a=file(nm,'w')
    a.write("$molecule\n0 1\n")
    a.write(xyz.replace(dist,str(r)))
    a.write("\n$end\n")
    a.write(rem)
    a.close()
    os.popen("qchem -nt 1 "+nmin+" "+nmout+" "+nmdir).read()
    op=file(nmout,'r').read()
    try:
        c=op.index("Total energy")
        return float(op[c:].splitlines()[0].split()[-1])
    except:
        print "Something went wrong for ",nm
        return
```

```python
print "This is to show how to run a miniature queue on your device"

import Queue
import threading

# input and output queues, maxsize is infinite when set to 0 or lower
q_in = Queue.Queue(maxsize=0)
q_out = Queue.Queue(maxsize=0)

# number of processors to use
num_worker_threads = 12

# process that each worker thread will execute until the Queue is empty
def worker():
    while True:
        # get item from queue, do work on it, let queue know processing is done for one item
        item = q_in.get()
        if len(item)==3:
            q_ser_run(item[0],item[1],dist=item[2])
        else:
            q_ser_run(item[0],item[1])
        q_out.put(item[:2])
        q_in.task_done()

# process that output queue calls
def printer():
    while True:
        # get an item processed by worker threads and print the result. Let queue know item has been processed
        item = q_out.get()
        #print "Completed ", item
        q_out.task_done()

# Initiate the input queue and pass function for processes to run
for i in range(num_worker_threads):
     t = threading.Thread(target=worker)
     t.daemon = True
     t.start()

# Launch output queue
t = threading.Thread(target=printer)
t.daemon = True
t.start()

# Fill the input queue
xr=arange(3.0,10.25,0.25)
for r in xr:
    q_in.put(("ne",r))
    q_in.put(("ne2",r,"3.089971792"))
    q_in.put(("ar",r))
    q_in.put(("ar2",r,"3.760016025"))

# Waiting for all queues to finish
q_in.join()
q_out.join()
```

```python
print "This performs the same as above in a more generalizable manner"

def qrun(name,mod,xyz='',ncpus=1):
    nm=name+"_"+str(mod)+".in"
    nmin=nm
    nmout=nm.replace(".in",".out")
    nmdir=nm.replace(".in",".dir")
    if os.path.isfile(nmout):
        op=file(nmout,'r').read()
        try:
            c=op.index("Total energy")
            return float(op[c:].splitlines()[0].split()[-1])
        except:
            print "Output not right"
            return
    a=file(nm,'w')
    a.write("$molecule\n0 1\n")
    a.write(xyz)
    a.write("\n$end\n")
    a.write(rem)
    a.close()
    os.popen("qchem -nt "+str(ncpus)+" "+nmin+" "+nmout+" "+nmdir).read()
    op=file(nmout,'r').read()
    try:
        c=op.index("Total energy")
        return float(op[c:].splitlines()[0].split()[-1])
    except:
        print "Something went wrong for ",nm
        return
    
def asz(val):
    return {
        'O': 16,
        'N': 14,
        'H': 1,
        'C': 12,
        'Ne': 20,
        'Ar': 40
        }.get(val, 0)
def dilate(name,m1,m2,lb,ub,inc):
    dimer=open(name+'.xyz','r').read().splitlines()[2:]
    for i in xrange(len(dimer)):
        dimer[i]=dimer[i].split()
    com1=[0.0,0.0,0.0]
    com2=[0.0,0.0,0.0]
    m1sum=0
    m2sum=0
    m1x=[]
    m1y=[]
    m1z=[]
    m2x=[]
    m2y=[]
    m2z=[]
    vector=[]
    d=0.00
    comvx=[]
    comvy=[]
    comvz=[]
    atoms1=[]
    atoms2=[]
    for i, row in enumerate(dimer):
        if i<m1:
            m1x.append(float(row[1]))
            m1y.append(float(row[2]))
            m1z.append(float(row[3]))
            atoms1.append(row[0])
            atsz=asz(row[0])
            com1[0]+=float(row[1])*atsz
            com1[1]+=float(row[2])*atsz
            com1[2]+=float(row[3])*atsz
            m1sum+=atsz
        else: #second monomer
            m2x.append(float(row[1]))
            m2y.append(float(row[2]))
            m2z.append(float(row[3]))
            atoms2.append(row[0])
            atsz=asz(row[0])
            com2[0]+=float(row[1])*atsz
            com2[1]+=float(row[2])*atsz
            com2[2]+=float(row[3])*atsz
            m2sum+=atsz
    for i in xrange(3): #finish weighting and generate center of mass vector and magnitude
        com1[i]/=m1sum
        com2[i]/=m2sum
        vector.append( com2[i]-com1[i] )
        d+=vector[i]*vector[i]
    for i in xrange(m2): #get displacements from center of mass coordinates
        comvx.append( m2x[i]-com2[0])
        comvy.append( m2y[i]-com2[1])
        comvz.append( m2z[i]-com2[2])
    d=math.sqrt(d)

    fragm1=""
    for i in xrange(m1):
		fragm1=fragm1+atoms1[i]+'\t'+str(m1x[i])+'\t'+str(m1y[i])+'\t'+str(m1z[i])+'\n'
        
    xyzs=[]
    for num in arange(lb,ub+inc,inc):
        xyz=fragm1
        newmx=[]
        newmy=[]
        newmz=[]
        for i in xrange(m2):
            newmx.append(com1[0]+comvx[i]+vector[0]*num/d)
            newmy.append(com1[1]+comvy[i]+vector[1]*num/d)
            newmz.append(com1[2]+comvz[i]+vector[2]*num/d)
        for i in xrange(m2):
            xyz=xyz+atoms2[i]+'\t'+str(newmx[i])+'\t'+str(newmy[i])+'\t'+str(newmz[i])+'\n'
        xyzs.append((num,xyz))
        #q_in.put((name,num,xyz))
        #qrun(name,num,xyz=xyz,ncpus=12)  #for running one at a time
    return xyzs
```

```python
print "Taking the above list for dissociating two fragments, we run a queue."

import Queue
import threading

# input and output queues, maxsize is infinite when set to 0 or lower
q_in = Queue.Queue(maxsize=0)
q_out = Queue.Queue(maxsize=0)

# number of processors to use
num_worker_threads = 12

# process that each worker thread will execute until the Queue is empty
def worker():
    while True:
        # get item from queue, do work on it, let queue know processing is done for one item
        item = q_in.get()
        if len(item)==3:
            qrun(item[0],item[1],xyz=item[2])
        else:
            qrun(item[0],item[1])
        q_out.put(item)
        q_in.task_done()

# process that output queue calls
def printer():
    while True:
        # get an item processed by worker threads and print the result. Let queue know item has been processed
        item = q_out.get()
        #print "Completed ", item
        q_out.task_done()

# Initiate the input queue and pass function for processes to run
for i in range(num_worker_threads):
     t = threading.Thread(target=worker)
     t.daemon = True
     t.start()

# Launch output queue
t = threading.Thread(target=printer)
t.daemon = True
t.start()

# Fill the input queue
for i in dilate("ne2",1,1,3.5,10.25,.25):
    q_in.put(("ne2",i[0],i[1]))
#    qrun("ne2",i[0],xyz=i[1])
# Waiting for all queues to finish
q_in.join()
q_out.join()
```

```python
print "Now let's do a 2-dimensional scan - setting up here"

def qrun(name,mod,xyz='',ncpus=1):
    nm=name+"_"+str(mod)+".in"
    nmin=nm
    nmout=nm.replace(".in",".out")
    nmdir=nm.replace(".in",".dir")
    if os.path.isfile(nmout):
        op=file(nmout,'r').read()
        try:
            c=op.index("Total energy")
            return float(op[c:].splitlines()[0].split()[-1])
        except:
            print "Output not right"
            return
    a=file(nm,'w')
    a.write("$molecule\n0 1\n")
    a.write(xyz)
    a.write("\n$end\n")
    a.write(rem)
    a.close()
    os.popen("qchem -nt "+str(ncpus)+" "+nmin+" "+nmout+" "+nmdir).read()
    op=file(nmout,'r').read()
    try:
        c=op.index("Total energy")
        return float(op[c:].splitlines()[0].split()[-1])
    except:
        print "Something went wrong for ",nm
        return

def trunc(f, n):
    '''Truncates/pads a float f to n decimal places without rounding'''
    return ('%.*f' % (n + 1,f))[:-1]

def shift(name,m1,m2,xl,xu,yl,yu,zl,zu,inc):
    dimer=open(name+'.xyz','r').read().splitlines()[2:]
    for i in xrange(len(dimer)):
        dimer[i]=dimer[i].split()
    com1=[0.0,0.0,0.0]
    com2=[0.0,0.0,0.0]
    m1sum=0
    m2sum=0
    m1x=[]
    m1y=[]
    m1z=[]
    m2x=[]
    m2y=[]
    m2z=[]
    vector=[]
    d=0.00
    atoms1=[]
    atoms2=[]
    for i, row in enumerate(dimer):
        if i<m1:
            m1x.append(float(row[1]))
            m1y.append(float(row[2]))
            m1z.append(float(row[3]))
            atoms1.append(row[0])
        else: #second monomer
            m2x.append(float(row[1]))
            m2y.append(float(row[2]))
            m2z.append(float(row[3]))
            atoms2.append(row[0])
    fragm1=""
    for i in xrange(m1):
		fragm1=fragm1+atoms1[i]+'\t'+str(m1x[i])+'\t'+str(m1y[i])+'\t'+str(m1z[i])+'\n'        
    xyzs=[]
    for x in arange(xl,xu+inc,inc):
        for y in arange(yl,yu+inc,inc):
            for z in arange(zl,zu+inc,inc):
                xyz=fragm1
                newmx=[]
                newmy=[]
                newmz=[]
                for i in xrange(m2):
                    newmx.append(m2x[i]+x)
                    newmy.append(m2y[i]+y)
                    newmz.append(m2z[i]+z)
                for i in xrange(m2):
                    xyz=xyz+atoms2[i]+'\t'+str(newmx[i])+'\t'+str(newmy[i])+'\t'+str(newmz[i])+'\n'
                xyzs.append(((trunc(x,1)+"_"+trunc(y,1)+"_"+trunc(z,1)),xyz))
    return xyzs
```
